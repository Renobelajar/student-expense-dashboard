4. Menampilkan Grafik

```python
import pandas as pd
import matplotlib.pyplot as plt

# 1. Load data
df = pd.read_csv('expense_data.csv')

# 2. Perbandingan Total: Income vs Expense
total_summary = df.groupby('Income/Expense')['Amount'].sum()

plt.figure(figsize=(12, 5))

# Subplot 1: Batang Income vs Expense
plt.subplot(1, 2, 1)
total_summary.plot(kind='bar', color=['red', 'green'])
plt.title('Total Income vs Expense')
plt.ylabel('Amount (INR)')
plt.xticks(rotation=0)

# 3. Perbandingan Pengeluaran per Kategori
# Kita filter hanya yang 'Expense' saja
expenses_only = df[df['Income/Expense'] == 'Expense']
cat_summary = expenses_only.groupby('Category')['Amount'].sum().sort_values(ascending=False)

# Subplot 2: Batang Kategori Pengeluaran
plt.subplot(1, 2, 2)
cat_summary.plot(kind='bar', color='skyblue')
plt.title('Pengeluaran Berdasarkan Kategori')
plt.ylabel('Amount (INR)')
plt.xticks(rotation=45)

plt.tight_layout()
plt.show()
```

```python
# Ringkasan total Income dan Expense
total_summary = df.groupby('Income/Expense')['Amount'].sum()

# Filter hanya Expense
expenses_only = df[df['Income/Expense'] == 'Expense']

# Total pengeluaran per kategori
cat_summary = expenses_only.groupby('Category')['Amount'].sum()

plt.figure(figsize=(12, 5))

# Diagram Lingkaran
plt.subplot(1, 2, 1)
plt.pie(
    total_summary,
    labels=total_summary.index,
    autopct='%1.1f%%',
    startangle=90
)
plt.title('Income vs Expense')

# Diagram Donut
plt.subplot(1, 2, 2)
plt.pie(
    cat_summary,
    labels=cat_summary.index,
    autopct='%1.1f%%',
    startangle=90,
    wedgeprops={'width': 0.7}
)
plt.title('Expense by Category')

plt.tight_layout()
plt.show()
```

5. Menentukan Pengeluaran Terbesar

```python
# Filter hanya Expense
expenses_only = df[df['Income/Expense'] == 'Expense']

# Total pengeluaran tiap kategori
cat_summary = expenses_only.groupby('Category')['Amount'].sum()

# Cari kategori dengan pengeluaran terbesar
largest_category = cat_summary.idxmax()
largest_amount = cat_summary.max()

print("Kategori Pengeluaran Terbesar :", largest_category)
print("Total Pengeluaran :", largest_amount)

# Visualisasi
plt.figure(figsize=(7, 5))

cat_summary.sort_values(ascending=False).plot(kind='bar')

plt.title('Pengeluaran Berdasarkan Kategori')
plt.ylabel('Amount (INR)')
plt.xticks(rotation=45)

# Tambahkan highlight pada pengeluaran terbesar
plt.axhline(largest_amount, linestyle='--')

plt.show()
```
