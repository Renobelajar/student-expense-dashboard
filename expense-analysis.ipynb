1. Menampilkan Dataset

Program harus dapat membaca dan menampilkan data pengeluaran.

```python
import pandas as pd

# Membaca data
df = pd.read_csv('expense_data.csv')

# Menampilkan data

df.sample(n=50, random_state=42)
```

2. Menghitung Total Pengeluaran

```python
# Menghitung total berdasarkan tipe
total = df.groupby('Income/Expense')['INR'].sum()

# Mengambil nilai spesifik
total_income = total['Income']
total_expense = total['Expense']

print(f"Total Pendapatan (Income)  : INR {total_income:,.2f}")
print(f"Total Pengeluaran (Expense): INR {total_expense:,.2f}")
print(f"-----------------------------------------")
print(f"Sisa Saldo                : INR {total_income - total_expense:,.2f}")
```
